```python
import matplotlib as mpl


def plot_errors(errors_df):
    # mpl.style.use('default')
    mpl.style.use('seaborn')

# 
    fig, ax = plt.subplots(figsize=(10,7))
    for i, para in enumerate(errors_df.index):
        ax.set_xlabel('relative $h_r= h / \\theta_i$ ')
        ax.set_ylabel('relative error')

        ax.scatter(errors_df.columns[:-2].astype(str), errors_df.loc[para][:-2]/errors_df.loc[para][-1], label=para)
        ax.plot(errors_df.columns[:-2].astype(str), errors_df.loc[para][:-2]/errors_df.loc[para][-1])

    ax.legend()
    plt.show()  
plot_errors(errors_df)
```
